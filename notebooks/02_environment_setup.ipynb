# 02 - Environment Setup

## Tujuan Notebook

Notebook ini digunakan untuk mendokumentasikan dan memvalidasi environment kerja project YouTube Rupiah Sentiment Analysis.

Tahap ini mencakup validasi:

1. Conda environment yang digunakan.
2. Versi Python dan lokasi executable Python.
3. Library utama project.
4. Struktur folder project.
5. File konfigurasi awal.
6. Keamanan file `.env`.
7. Status awal Git repository.

Notebook ini tidak digunakan untuk melakukan crawling data. Crawling komentar YouTube akan dilakukan pada notebook berikutnya, yaitu `03_youtube_comment_crawling.ipynb`.

```python
# Mengecek versi Python dan lokasi executable Python
# Tujuannya untuk memastikan notebook berjalan pada environment project yang benar.

import sys
import platform

print("Python Version:")
print(sys.version)

print("\nPython Executable:")
print(sys.executable)

print("\nOperating System:")
print(platform.platform())
```

```python
# Mengecek lokasi notebook dan root folder project

from pathlib import Path

CURRENT_DIR = Path.cwd()
PROJECT_ROOT = CURRENT_DIR.parent

print("Current Directory:")
print(CURRENT_DIR)

print("\nProject Root:")
print(PROJECT_ROOT)

print("\nApakah folder project root valid?")
print(PROJECT_ROOT.exists())
```

```python
# Mengecek library utama yang akan digunakan dalam project

import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
import sklearn
import nltk
import Sastrawi
import streamlit
import plotly
import joblib
import yaml
import tqdm
import emoji

print("Semua library utama berhasil diimpor.")
print("Pandas:", pd.__version__)
print("NumPy:", np.__version__)
print("Scikit-learn:", sklearn.__version__)
print("Matplotlib:", matplotlib.__version__)
print("Seaborn:", sns.__version__)
print("Streamlit:", streamlit.__version__)
```

```python
# Mengecek apakah struktur folder utama project sudah tersedia

required_folders = [
    "notebooks",
    "data",
    "data/raw",
    "data/interim",
    "data/processed",
    "data/labeled",
    "data/sample",
    "src",
    "models",
    "outputs",
    "outputs/figures",
    "outputs/reports",
    "streamlit_app",
    "docs"
]

print("Validasi Struktur Folder Project:\n")

for folder in required_folders:
    folder_path = PROJECT_ROOT / folder
    status = "OK" if folder_path.exists() else "Tidak ditemukan"
    print(f"{folder:<25} : {status}")
```

```python
# Mengecek apakah file konfigurasi utama sudah tersedia

required_files = [
    ".env",
    ".env.example",
    ".gitignore",
    ".gitattributes",
    "requirements.txt",
    "environment.yml",
    "README.md",
    "config.yaml",
    "src/__init__.py",
    "streamlit_app/app.py",
    "streamlit_app/utils.py"
]

print("Validasi File Konfigurasi Project:\n")

for file in required_files:
    file_path = PROJECT_ROOT / file
    status = "OK" if file_path.exists() else "Tidak ditemukan"
    print(f"{file:<30} : {status}")
```

```python
# Mengecek apakah file .env tersedia dan berisi variabel YOUTUBE_API_KEY
# API key tidak ditampilkan untuk menjaga keamanan credential.

import os
from dotenv import load_dotenv

env_path = PROJECT_ROOT / ".env"

load_dotenv(dotenv_path=env_path)

youtube_api_key = os.getenv("YOUTUBE_API_KEY")

print("File .env ditemukan:", env_path.exists())

if youtube_api_key:
    print("YOUTUBE_API_KEY tersedia di file .env.")
    print("Status keamanan: API key tidak ditampilkan di notebook.")
else:
    print("YOUTUBE_API_KEY belum ditemukan. Periksa kembali isi file .env.")
```

```python
# Membaca file config.yaml untuk memastikan konfigurasi project dapat digunakan

import yaml

config_path = PROJECT_ROOT / "config.yaml"

with open(config_path, "r", encoding="utf-8") as file:
    config = yaml.safe_load(file)

print("Config berhasil dibaca.")
print("Nama Project:", config["project"]["name"])
print("Judul Project:", config["project"]["title"])
print("Target minimal clean comments:", config["youtube"]["min_clean_comments_target"])
print("Target maksimal crawling comments:", config["youtube"]["max_comments_target"])
```

```python
# Mengecek apakah .gitignore sudah memuat aturan untuk mengabaikan file .env

gitignore_path = PROJECT_ROOT / ".gitignore"

with open(gitignore_path, "r", encoding="utf-8") as file:
    gitignore_content = file.read()

if ".env" in gitignore_content:
    print("File .env sudah terdaftar di .gitignore.")
    print("Status: Aman, .env tidak seharusnya masuk ke GitHub.")
else:
    print("Peringatan: .env belum ditemukan di .gitignore.")
```
